## Transformadas de Fourier

El apartado anterior trataba de las series de Fourier y de como una funcin peridica se poda descomponer en una suma infinita de funciones sinusoidales. Ahora bien, nos encontramos con el siguiente problema, qu pasa si la funcin en cuestin no es peridica?


### Transformada continua de Fourier

Para esto se define el concepto de transformada de Fourier. Imaginemos una funcin escaln, como se sabe no peridica, sin embargo, para convertirla en una funcin con periodicidad solo hace falta copiar este escaln y repartirlo infinitamente por el eje X y cada cierto perdiodo. Hemos convertido nuestra funcin en peridica aunque de poco nos vale puesto que no es nuestra funcin original y la representacin de Fourier no sera la que buscamos. Lo que podemos hacer es dejar que el periodo de dicha funcin inventada tienda a infinito y ver que pasa. Primeramente encontramos que la frecuecia tiende a cero por lo que creamos una representacin continua de nuestra funcin en un espacio frecuencial. Por otro lado, si cogemos la definicin de la serie de Fourier en su representacin complejo-exponencial y sustituimos el valor de los coeficientes dentro de la intergral, tenemos:

$$
f(t)=\sum_{n=-\infty}^{\infty}\left[\frac{1}{T} \int_{-T / 2}^{T / 2} f(t) e^{-j n \omega_0 t} d t\right] e^{j n \omega_0 t}
$$

Por otro lado, el espaciado entre los armnicos es de $\Delta \omega=(n+1) \omega_0-n \omega_0=\omega_0=\frac{2 \pi}{T}$. Si sustituimos arriba

$$
f(t)=\frac{1}{2 \pi} \sum_{n=-\infty}^{\infty}\left[\int_{-T / 2}^{T / 2} f(t) e^{-j n \omega_0 t} d t\right] \Delta \omega e^{j n \omega_0 t}
$$

Ahora tendremos en cuenta las siguientes consideraciones. Si dejamos que $T \rightarrow \infty$, la suma se convierte integral, el incremento $\Delta \omega$ se convierte en un diferencial $d \omega$, y la frecuencia discreta $n \omega_0$ se convierte en una frecuencia continua $\omega$ [6]. Por lo tanto, como $T \rightarrow \infty$:

$$
\begin{aligned}
\sum_{n=-\infty}^{\infty} & \Rightarrow \int_{-\infty}^{\infty} \\
\Delta \omega & \Rightarrow d \omega \\
n \omega_0 & \Rightarrow \omega
\end{aligned}
$$

De esta manera tenemos que la representacin en serie de senos y cosenos de la funcin, ha quedado como una representacin integral de exponenciales complejas
$$
f(t)=\frac{1}{2 \pi} \int_{-\infty}^{\infty}\left[\int_{-\infty}^{\infty} f(t) e^{-j \omega t} d t\right] e^{j \omega t} d \omega
$$

De donde se extrae el famoso valor de la transformada de Fourier:
$$
F(\omega)=\mathcal{F}[f(t)]=\int_{-\infty}^{\infty} f(t) e^{-j \omega t} d t
$$

Los resultados y las aplicaciones de esta transformada son mltiples. Uno de los resultados es que si tenemos, por ejemplo, la intensidad que recorre una malla de un circuito, esta se define en el dominio del tiempo (DT), mientras que la transformada de la intensidad se define en el dominio de la frecuencia (DF)
