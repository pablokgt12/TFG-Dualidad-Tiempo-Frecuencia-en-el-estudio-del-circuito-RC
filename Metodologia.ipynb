## Carga y descarga del condensador
### Introducción básica

El condensador es un dispositivo compuesto generalmente por dos conductores, que adquieren cargas superficiales absolutas iguales pero de signo contrario, y un material dieléctrico o aislante entre ellos creándose una diferencia de potencial entre los mismos. Los conductores pueden ser placas plano-paralelas, esféricos, cilíndricos, etc... En el material dieléctrico se encuentra un campo producido por los conductores cargados, mientras que en el interior de estos últimos el campo es nulo. Esto se debe a que los conductores tienen
electrones libres, que consiguen oponerse al campo hasta que resulta ser nulo, en cambio, los dieléctricos no tienen electrones libres, sino que se polarizan enviando a un lado la carga positiva y al otro la positiva creando dipolos que no pueden neutralizar el campo eléctrico en el interior de éstos. 

Por otro lado, el proceso de carga y descarga de un capacitor es un fenómeno fundamental en la teoría de circuitos y la electrónica, que encuentra una amplia gama de aplicaciones en numerosos campos, desde la electrónica básica hasta la tecnología de vanguardia, los comentaremos a continuación. Este proceso implica la transferencia de carga eléctrica hacia y desde un capacitor, un componente pasivo que tiene la capacidad de almacenar energía en forma de campo eléctrico entre dos placas conductoras separadas por un dieléctrico.

### Aplicaciones

   Filtros de señales: Los circuitos RC (resistencia-capacitancia) se utilizan comúnmente en la electrónica para filtrar señales. La carga y descarga del capacitor en estos circuitos permite filtrar frecuencias no deseadas de una señal eléctrica.

   Fuentes de energía: Los capacitores se utilizan en aplicaciones de almacenamiento de energía, como en circuitos de temporización, fuentes de alimentación ininterrumpida (UPS) y en sistemas de energía renovable para suavizar la salida de energía.

   Memoria: En la electrónica digital, los capacitores se utilizan en circuitos de memoria volátil, como en RAM (memoria de acceso aleatorio), donde se cargan y descargan para representar datos binarios.

   Arranque de motores: En aplicaciones de motores eléctricos, los capacitores se utilizan para proporcionar un impulso inicial de energía para iniciar el motor.

   Circuitos de temporización: Los circuitos RC se utilizan para generar retardos de tiempo precisos en aplicaciones como temporizadores, osciladores y pulsadores.

#### Carga del condensador
La ecuación diferencial para un circuito serie R-C es:
$$
v_e(t)=v_R(t)+v_C(t)=R C \frac{d v_C(t)}{d t}+v_C(t)
$$

La solución general de la homogénea es:

$$
\left.v_c(t)\right|_{g h}=A e^{-\frac{t}{\tau}}
$$
donde $\tau=\mathrm{RC}$, es la constante de tiempo del sistema y A es la constante de integración.

Para obtener la solución particular hay que especificar el tipo de alimentación que aplicamos. También hay que especificar las condiciones iniciales del circuito para $t<0$. La señal de entrada viene definida por

$$
v_e(t)= \begin{cases}0 & \text { si } t \epsilon[-\infty, 0[ \\ V_0 & \text { si } t \epsilon\left[0, t_w[\right. \\ 0 & \text { si } t \epsilon\left[t_w, \infty[\right.\end{cases}
$$

Se tiene que

$$
v_c(t)=\frac{1}{C} \int_{-t}^t i(\alpha) d \alpha
$$

donde - $t$ es un instante menor que cero. Vemos que si la intensidad que recorre el circuito está acotada, la función $v_C(t)$ es continua.
La solución particular de la no homogénea para t $>0$ se obtiene de forma inmediata por inspección

$$
v_c(t)= \begin{cases}V_0 & \text { si } t \epsilon\left[0, t_w[\right. \\ 0 & \text { si } t \epsilon\left[t_w, \infty[\right.\end{cases}
$$

y la solución final, para una señal de entrada tipo pulso de altura V0 y anchura tw, vendrá dada por:

$$
v_c(t)=V_0\left[1-e^{-\left(\frac{t}{\tau}\right)}\right] \forall t \epsilon\left[0, t_w[\right.
$$

La intensidad que recorre el circuito puede calcularse sobre la resistencia mediante la Ley de Ohm y vendrá dada por

$$
i(t)= \begin{cases}0 & \text { si } t \epsilon[-\infty, 0[ \\ \frac{V_0}{R} e^{-\left(\frac{t}{\tau}\right)} & \text { si } t \epsilon\left[0, t_w[\right.\end{cases}
$$

y la caída de tensión en la resistencia

$$
v_R(t)= \begin{cases}0 & \text { si } t \epsilon[-\infty, 0[ \\ V_0 e^{-\left(\frac{t}{\tau}\right)} & \text { si } t \epsilon\left[0, t_w[\right.\end{cases}
$$

Vemos que $i(t)$ no es continua, pero no hay ninguna razón por la que deba serlo.

#### Descarga del condensador
En $t=t_w$ hacemos que $v_e(t)=0$. La caída de tensión en el con densador satisface una ecuación diferencial homogénea [4]

$$
0=R C \frac{d v_c(t)}{d t}+v_c(t)
$$

La solución es análoga a la del proceso de carga, pero con un valor de la constante distinto

$$
v_c\left(t_w\right)=V_0
$$

$$
v_c(t)=B e^{-\left(\frac{t}{\tau}\right)}
$$

El valor $V_0$ es un valor asintótico que se alcanza después de un tiempo infinito. En la práctica se tomará el valor de carga en que se detenga el proceso. La solución para $t>t_w$ es
- Si el proceso de carga se hubiese completado

$$
v_c(t)=V_0 e^{-\left(\frac{t-t_w}{\tau}\right)}
$$

- Si el proceso de carga no se hubiese completado tendríamos una solución dada por

$$
v_c(t)=V_0\left(e^{\frac{t_u}{\tau}}-1\right) e^{-\left(\frac{t}{\tau}\right)}
$$

Por lo tanto, la intensidad que recorre el circuito y la caída de tensión en la resistencia vienen dadas por
- Si el proceso de carga se hubiese completado

$$
\begin{gathered}
i(t)=-\frac{V_0}{R} e^{-\left(\frac{t-t_w}{\tau}\right)} \\
v_R(t)=-V_0 e^{-\left(\frac{t-t_w}{\tau}\right)}
\end{gathered}
$$

- Si el proceso de carga no se hubiese completado
$$
\begin{gathered}
i(t)=-\frac{V_0}{R}\left(e^{\frac{t_w}{\tau}}-1\right) e^{-\left(\frac{t}{\tau}\right)} \\
v_R(t)=-V_0\left(e^{\frac{t_w}{\tau}}-1\right) e^{-\left(\frac{t}{\tau}\right)}
\end{gathered}
$$

## Respuesta en frecuencia de un cicuito RC

Pasando ahora a la segunda práctica, nos cenctraremos en la respuesta frecuencia del mismo circuito. Para ello trabajaremos en el dominio de las frecuencias. Desde el punto de vista físico ambos sistemas que estamos comparando son similares, constante de una fuente de corriente, un condensador y una resistencia. La diferencia reside en el tipo de señal de la fuente de alimentación, en el primero era una fuente de corriente continua y ahora es alterna. El proceso de toma de medidas en el laboratorio es muy distinto el uno del otro, en el primero se lee de un voltímetro para cada intervalo de tiempo, mientras que ahora se trata de leer en un osciloscopio el voltaje, que al tratarse de una onda sinusoidal, consta de magnitud y de fase. Aquí, es donde se pone muy interesante el problema que hemos planteado. La comparativa de las dos prácticas resulta intuitiva cuando pensamos que son las dos caras de una misma moneda, y es la transformada de Fourier la que las conecta [5]. Por lo tanto, es posible pasar de mediciones de voltaje (corriente continua), a mediciones de magnitud y fase (corriente alterna) mediante dicha trasnformada.

Si dejamos constante la amplitud de la fuente sinusoidal y variamos la frecuencia, obtenemos la respuesta en frecuencia del circuito. La respuesta en frecuencia puede considerarse como una descripción completa de el comportamiento sinusoidal en estado estacionario de un circuito en función de la frecuencia [6].

### Fundamento teórico

La respuesta en frecuencia de un circuito es la variación de su comportamiento con cambio en la frecuencia de la señal [6].

Considerando de nuevo el mismo circuito salvo que modificando el tipo de corriente que se suministra, trabajaremos con la siguiente configuaración circuital:



Si aplicamos la segunda ley de Kirchhoff en la malla nos queda que:
$$
\begin{gathered}
V_e(t)-V_c(t)-V_r(t)=0 \\
V_e(t)=\frac{1}{C} \int i(t) d t+R i(t)
\end{gathered}
$$

Utilizando la definición de derivada, nos queda que, en el dominio del tiempo:
$$
V_e=R C \frac{\partial v_c(t)}{\partial t}+v_c(t)
$$

Que es una ecuación diferencial de primer orden con coeficientes constantes.
Si ahora desarollamos lo que vale $V_c$ y $V_e$ y los llevamos al dominio de las frencuencias podemos considerar que:

- $V_e=v_0 \cos (\omega t+\phi)=\operatorname{Re}\left[v_0 e^{j(\omega t+\phi)}]=\operatorname{Re}\left[V_0 e^{j \omega t}\right]\right.$
- $V_c=v_c \operatorname{cost}(\omega t+\gamma)=\operatorname{Re}\left[v_c e^{j(\omega t+\gamma)}\right]=\operatorname{Re}\left[V_c e^{j \omega t}\right]$

Sustituyendo los valores en forma fasorial en la ecuación diferencial nos queda, finalmente que:

$$
V_0=(j \omega R C+1) V_c \Rightarrow V_c=\frac{V_0}{j \omega R C+1}
$$

Tomando módulos:

$$
\begin{aligned}
& \left|V_c\right|=\frac{v_o}{\sqrt{(\omega R C)^2+1}} \\
& \gamma=\phi-\arctan (\omega R C)
\end{aligned}
$$

Luego ya tenemos las características de la tensión que cae en el condensador, que tiene la forma:

$$
v_c(t)=\left|V_c\right| \cos (\omega t+\gamma)
$$

Para el resto de variables circuitales, como la intensidad, sólo nos queda desarrollar la expresión:

$$
i(t)=C \frac{d v_c}{d t}=C \frac{d}{d t}\left[\left|V_c\right| \cos (\omega t+\gamma)\right]=-\left|V_c\right| C \omega \operatorname{cosen}(\omega t+\gamma)=\left|V_c\right| C \omega \cos (\omega t+\gamma+\pi / 2)
$$

Finalmente, a partir de las relaciones anteriores podemos definir la función de transferencia del circuito con salida en el condensador, que definiremos en este caso como:

$$
T_c \equiv \frac{V_c}{V_e}=\frac{1}{j \omega R C+1}
$$

Esta función nos brindará, conocidos los valores de R y C, la tensión que cae en el condensador a una determinada frecuencia $\omega$ de la siguiente forma, y considerando los fasores definidos anteriormente:

$$
V_c=T_c V_e=\frac{V_0 e^{j \phi}}{R \omega C+1}=\frac{v_0}{\sqrt{(\omega R C+1)^2}} e^{j(\phi-\arctan (\omega R C))}
$$

Llevando la expresión al dominio del tiempo:

$$
v_c(t)=\frac{v_0}{\sqrt{1+(\omega R C)^2}} \cos [\omega t+\phi-\operatorname{arctg}(\omega R C)]
$$

Luego, conociendo la función de transferencia podemos obtener el valor de las tensiones a cualquier frecuencia. Si llevamos a cabo el mismo desarrollo pero con salida en la resistencia, nos queda que la función $T_r(\omega)$ :

$$
T_R=\frac{V_R}{V_e}=\frac{j \omega R C}{j \omega R C+1}=\frac{\omega R C}{\sqrt{1+(\omega R C)^2}} e^{j\left(-\arctan (\omega R C)+\frac{\pi}{2}\right)}
$$

#### Diagramas de Bode

Como íbamos diciendo, las respuestas frecuenciales sinusoidales en estado estacionario de los circuitos son de importancia en muchas aplicaciones, especialmente en comunicaciones y sistemas de control. Una aplicación específica son los filtros eléctricos que bloquean o eliminan señales con frecuencias no deseadas y pasan señales de frecuencias deseadas. Los filtros se utilizan en sistemas de radio, televisión y telefonía para separar una frecuencia de emisión de otra.

Dicho esto, analizaremos el concepto de función de transferencia de un modo más abstracto. La función de transferencia H(w) de un circuito es la relación, en función de la frecuencia
de un fasor de salida Y(w) (tensión o corriente de un elemento) a un fasor de de entrada X(w) (tensión o corriente de la fuente)

![h w.png](<attachment:h w.png>)

Un fasor es una representación matemática utilizada para analizar señales. Simplifica el análisis de circuitos de alterna representando la amplitud y la fase de la tensión como un número complejo.

Por lo tanto 

$$
\mathbf{H}(\omega)=\frac{\mathbf{Y}(\omega)}{\mathbf{X}(\omega)}
$$

Suponiendo condiciones iniciales nulas. Dado que la entrada y la salida pueden ser tensión o corriente en cualquier punto del circuito, hay cuatro posibles funciones de transferencia

$$
\begin{aligned}
& \mathbf{H}(\omega)=\text { Voltage gain }=\frac{\mathbf{V}_o(\omega)}{\mathbf{V}_i(\omega)} \\
& \mathbf{H}(\omega)=\text { Current gain }=\frac{\mathbf{I}_o(\omega)}{\mathbf{I}_i(\omega)} \\
& \mathbf{H}(\omega)=\text { Transfer Impedance }=\frac{\mathbf{V}_o(\omega)}{\mathbf{I}_i(\omega)} \\
& \mathbf{H}(\omega)=\text { Transfer Admittance }=\frac{\mathbf{I}_o(\omega)}{\mathbf{V}_i(\omega)}
\end{aligned}
$$

Y como vemos, si particularizamos para resistencia o condensador, recuperamos las ecuaciones que se plantearon anteriormente.

Pasemos a los diagramas de Bode propiamente dichos. Los diagramas de Bode son dos gráficos, el primero semilogarítmicos de la magnitud (en decibelios) y el segundo la fase
(en grados) de una función de transferencia, ambos en función de la frecuencia.[6]

Los gráficos de Bode contienen la misma información que los gráficos no logarítmicos analizados anteriormente, pero son mucho más fáciles de construir, como veremos en breve. La función de transferencia se puede escribir como

$$
\mathbf{H}=H \not \phi=H e^{j \phi}
$$

Tomando logaritmos naturales

$$
\ln \mathbf{H}=\ln H+\ln e^{j \phi}=\ln H+j \phi
$$

Por lo que la parte real $\ln \mathbf{H}$ es una función de la magnitud y la imaginaria de la fase. La ganancia en un diagrama de Bode es:

$$
H_{\mathrm{dB}}=20 \log _{10} H
$$

## Teorema de muestreo y de la frecuencia de Nyquist 

De manera introductoria, citaremos una de las formulaciones más conocidas del teorema y es la que sigue: "Para poder reconstruir una señal muestreada, la frecuencia de muestreo debe ser superior al doble del ancho de banda."[7]

### Concepto de muestreo

Empecemos diciendo que las señales físicas en el mundo real son, en su mayoria, analógicas. Por lo que si imaginamos una señal del mundo real, será continua y que si queremos digitalizarla, necesitaremos de memoria infinita para almacenar el infinito grado de detalle que tienen estas señales. Para solventar este problema, la muestreamos, es decir, cada cierto inveravalo de tiempo, medimos, esa será nuestra señal digital. Aspectos intuitivos de este proceso es que, cuanto menor sea el intervalo de tiempo en el que se mida, en adelante periodo de muestreo, mayor será la concordancia entre nuestra señal real y muestreada, pero también mayor será la memoria necesaria para almacenarla. Por lo tanto hay que encontrar un balance en el periodo de muestreo.[7]

### Muestreo de un tren de pulsos y condiciones para la recuperación de la señal original

Una vez que hemos muestreado nuestra señal, hemos conseguido digitalizarla y por tanto ahora podemos trabajar con ella, pero nos preguntamos si con esos puntos que se obtienen al muestrear es posible recuperar nuestra señal. Pues veremos a continuación que si nuestra señal cumple ciertos requisitos, sí será posible. Esta condición es que nuestra señal es debe ser de banda limitada y nuestras muestras lo suficientemente próximas.

El concepto de tren de pulsos es claro, vamos a multiplicar nuestra señal real (no muestreada), por un tren de deltas. En este proceso estamos perdiendo la información de lo que le pasaba a la señal real por lo que si somos capaces de con dicha señal resultante, recuperar mi señal original, demostraré la posibilidad del proceso y sentaré las bases de las condiciones necesarias para que esto ocurra [8].



Las ecuaciones que planteamos para reproducir el comportamiento de esta figura son:

$$
\begin{aligned}
& x_p(t)=x(t) p(t) \\
& p(t)=\sum_{n=-\infty}^{+\infty} \delta(t-n T) \\
& x_p(t)=\sum_{n=-\infty}^{+\infty} x(n T) \delta(t-n T)
\end{aligned}
$$

Y en el dominion de la frecuencia:

$$
\begin{aligned}
&X_p(f)=[X(f) * P(f)]\\
&\begin{aligned}
& P(f)=\frac{1}{T} \sum_{k=-\infty}^{+\infty} \delta\left(w-k f_s\right) \\
& X_p(f)=\frac{1}{T} \sum_{k=-\infty}^{+\infty} X\left(w-k f_s\right)
\end{aligned}
\end{aligned}
$$

El espectro en $f$ de la señal muestreada $x_p(t)$, es el espectro de la señal original $X(w)$ desplazado en múltiplos de la frecuencia de muestreo $\mathrm{w}_{\mathrm{s}}=1 / \mathrm{T}$



Vemos que el efecto del muestreo ideal sobre el espectro en frecuencias original, es repetirlo alrededor de la frecuencia de muestreo $\mathrm{f}_{\mathrm{S}}$.

Si disminuimos el valor de $f_S$ los espectros se empiezan a acercar y no hay problema hasta $f_S-f_M=f_M$

Como vemos en el espectro $X_p(f)$ de la figura anterior, podremos recuperar $X(f)$ original siempre y cuando $f_S-f_M>f_M$ y los espectros repetidos no se solapen. En efecto si disminuimos $f_S$ hasta llegar a un punto en que se mezclen, no será posible recuperar el espectro original.

Condiciones para reconstruir la señal original usando una digitalizada: 

-Si $x(t)$ es de banda limitada con $X(f)=0$ para $|f|>f_M$ entonces $x(t)$ está unívocamente determinada por sus muestras $x(n T)$, $\mathrm{n}=0, \pm 1, \pm 2 \ldots . . \mathrm{Si}_{\mathrm{s}}>2 \mathrm{f}_{\mathrm{M}}$.

-Para recuperar la señal original, se procesan las muestras con un filtro pasabajos ideal, con $f$ de corte $>f_M$ y menor que $f_s-f_M$.

-A $\mathrm{f}_{\mathrm{s}}$ se la conoce como la frecuencia de Nyquist

En conclusión se extrae la siguiente información:

$$
\begin{aligned}
& f_s > 2 f_M \\
& f_M={T} \\
& f_s<\frac{2}{T}
\end{aligned}
$$

Que también lo podemos expresar como:

$$
\begin{aligned}
& T<\frac{\pi}{\omega_m}
\end{aligned}
$$

Que se conoce como el teorema de muestreo de Nyquist.

A modo de resumen de como obtener la señal original a traves de la señal muestreada. Basta con que se aplica un filtro paso-baja al espectro de la señal muestreada y amplificar el resultado por un factor T.

## Análisis de Fourier

El análisis de Fourier es una herramienta matemática que se utilizar para descomponer funciones o señales complejas del mundo real en sus componentes más fundamentales, a menudo funciones sinusoidales, mediante un proceso denominado transformada de Fourier [9].

En esencia, el análisis de Fourier lidia con la representación de funciones complejas que se descomponen en sumas de funciones sinusoidales (seno y coseno). La idea del desarrollo que hay detrás es que muchas funciones periódicas, pueden descomponerse en una suma infinita de funciones seno y coseno de amplitud y frecuecia variable. Esta descomposición nos permite comprender mejor el comportamiento de los sistemas en los que muchas señales complejas están involucradas, analizando sus componentes de frecuencia.

Atendiendo al tipo de transformadas que se haga, podemos distinguir dos análisis de Fourier:

- El primero, es el análisis de Fourier continuo, se utiliza en el mundo más abstracto, y en el que se quiere llegar a resultados teóricos puros. Involucra integrales.
- Por otro lado, tenemos el análisis de Fourier discreto (DFT), es el que utilizaremos en el desarrollo de estas dos prácticas de laboratorio, puesto que es el más adecuado para obtener resultados computaciones de funciones discretas. La DFT crea una sucesión finita de muestras equi-espaciadas de una función, en su correspondiente secuencia de números complejos, que representan la amplitud y la fase de ondas sinusoidales a frecuencias discretas. Posteriormente, se realizará un análisis en profundidad del análisis discreto de Fourier y de dicha transformada, la DFT.


### Series de Fourier

Para entender el contexto del análisis de Fourier y todas sus connotaciones, es necesario explicar conceptos tales como, series trigonométricas de Fourier, espectro en frecuencia o simetría. Vamos con ello:

Mientras Fourier estudiaba el flujo de calor, descubrió que su comportamiento se regía por una función no sinusoidal pero periódica. De acuerdo con el teorema de Fourier, cualquier función periódica puede expresarse en términos de una suma infinita de senos y cosenos que son múltiplos de la frecuencia de la función, $\omega_0$. Expresando este resultado matemáticamente, tenemos:

$$
\begin{aligned}
f(t)= & a_0+a_1 \cos \omega_0 t+b_1 \sin \omega_0 t+a_2 \cos 2 \omega_0 t \\
& +b_2 \sin 2 \omega_0 t+a_3 \cos 3 \omega_0 t+b_3 \sin 3 \omega_0 t+\cdots
\end{aligned}
$$
o
$$
f(t)=\underbrace{a_0}_{\mathrm{C.D}}+\underbrace{\sum_{n=1}^{\infty}\left(a_n \cos n \omega_0 t+b_n \sin n \omega_0 t\right)}_{\mathrm{C.C}}
$$

C.D Y C.C hace referencia a los tipos de corriente.

"Donde $\omega_0=2 \pi / T$ se denomina frecuencia fundamental. La función sinoidal $\sin n \omega_0 t$ o $\cos n \omega_0 t$ se llama el enésimo armónico de $f(t)$; puede ser par o impar dependiendo de la naturaleza de la función, de su simetría. Esta ecuación se llama serie trigonométrica de Fourier de $f(t)$. Las constantes $a_n$ y $b_n$ son los coeficientes de Fourier. El coeficiente $a_0$ es la componente C.C o el valor medio de $f(t)$. Los coeficientes $a_n$ y $b_n$ son las amplitudes de las sinusoides en la componente C.A" [6].

Resumiendo lo que es una serie trigonométrica de Fourier, podemos decir que es otra forma de expresar una función periódica como suma infinita de senos y cosenos.

Decimos que una función es de Fourier, si cumple cada uno de los siguientes requisitos:

- La función periódica es inyectiva.
- Si tiene discontinuidades en un periodo, son un número finito de discontinuidades.
- Está acotada, tiene un número finito de máximos y mínimos en cualqueir periodo.
- La integral $\int_{t_0}^{t_0+T}|f(t)| d t<\infty$ para cualquier $t_0$

A estas condiciones se las conoce como de Dirichlet, aun que no son condiciones necesarias, son suficientes para que exista la correspondiente serie de Fourier [6].

No es complicado calcular el valor de los coeficientes de la serie de Fourier, son los siguientes:

$$
a_0=\frac{1}{T} \int_0^T f(t) d t
$$

Donde $a_{0}$ se le conoce como el valor medio de la función. Puesto que las funciones sinusoidales tienen valor medio nulo. Los coeficientes de Fourier se calculan:

$$
a_n=\frac{2}{T} \int_0^T f(t) \cos n \omega_0 t d t
$$

$$
b_n=\frac{2}{T} \int_0^T f(t) \sin n \omega_0 t d t
$$

Y que apicando las relaciones matemáticas de las identidades trigonométricas, podemos expersar todos estos resultados en forma de amplitud y fase:

$$
f(t)=a_0+\sum_{n=1}^{\infty} A_n \cos \left(n \omega_0 t+\phi_n\right)
$$

$$
A_n=\sqrt{a_n^2+b_n^2}, \quad \phi_n=-\tan ^{-1} \frac{b_n}{a_n}
$$

Del mismo modo, utilizando la representación del seno y coseno como funciones exponenciales complejas (Identidad de Euler), podemos expresar los resultados anteriores en su representación exponencial compleja:

$$
f(t)=\sum_{n=-\infty}^{\infty} c_n e^{j n \omega_0 t}
$$

$$
c_n=\frac{1}{T} \int_0^T f(t) e^{-j n \omega_0 t} d t
$$

Con estos últimos resultados, estamos listos para definir uno de los términos que habiamos comentado anteriormente. Es de una importancia notoria en el desarrollo de esta práctica, puesto que los desarrollos gráficos que se prevén crear tanto por nosotros mismos como por el alumnado en prácticas de laboratorio futuras se usará, y es el espectro en frecuencias.

El espectro de frecuencias de una señal consiste en graficar las amplitudes y fases de los armónicos de Fourier en función de la frecuencia [6].

Otro punto relevante a analizar es el de la simetría, pues nos permite, con un analisis visual de la función, discernir que coeficientes de la serie de Fourier son cero. Si la funcion es par, es decir, $f(t)=f(-t)$, los coeficientes $b_n$ son automáticamente todos nulos, puesto que se trata de integrar una función par en un intervalo impar. Consecuentemente, si la función es impar, es decir, $f(t)=-f(-t)$, los coeficientes $a_n$ serán todos cero. Por el motivo recíproco, integral de una función impar en un intervalo par.

## Transformadas de Fourier

El apartado anterior trataba de las series de Fourier y de como una función periódica se podía descomponer en una suma infinita de funciones sinusoidales. Ahora bien, nos encontramos con el siguiente problema, ¿qué pasa si la función en cuestión no es periódica?


### Transformada continua de Fourier

Para esto se define el concepto de transformada de Fourier. Imaginemos una función escalón, como se sabe no periódica, sin embargo, para convertirla en una función con periodicidad solo hace falta copiar este escalón y repartirlo infinitamente por el eje X y cada cierto perdiodo. Hemos convertido nuestra función en periódica aunque de poco nos vale puesto que no es nuestra función original y la representación de Fourier no sería la que buscamos. Lo que podemos hacer es dejar que el periodo de dicha función inventada tienda a infinito y ver que pasa. Primeramente encontramos que la frecuecia tiende a cero por lo que creamos una representación continua de nuestra función en un espacio frecuencial. Por otro lado, si cogemos la definición de la serie de Fourier en su representación complejo-exponencial y sustituimos el valor de los coeficientes dentro de la intergral, tenemos:

$$
f(t)=\sum_{n=-\infty}^{\infty}\left[\frac{1}{T} \int_{-T / 2}^{T / 2} f(t) e^{-j n \omega_0 t} d t\right] e^{j n \omega_0 t}
$$

Por otro lado, el espaciado entre los armónicos es de $\Delta \omega=(n+1) \omega_0-n \omega_0=\omega_0=\frac{2 \pi}{T}$. Si sustituimos arriba

$$
f(t)=\frac{1}{2 \pi} \sum_{n=-\infty}^{\infty}\left[\int_{-T / 2}^{T / 2} f(t) e^{-j n \omega_0 t} d t\right] \Delta \omega e^{j n \omega_0 t}
$$

Ahora tendremos en cuenta las siguientes consideraciones. Si dejamos que $T \rightarrow \infty$, la suma se convierte integral, el incremento $\Delta \omega$ se convierte en un diferencial $d \omega$, y la frecuencia discreta $n \omega_0$ se convierte en una frecuencia continua $\omega$ [6]. Por lo tanto, como $T \rightarrow \infty$:

$$
\begin{aligned}
\sum_{n=-\infty}^{\infty} & \Rightarrow \int_{-\infty}^{\infty} \\
\Delta \omega & \Rightarrow d \omega \\
n \omega_0 & \Rightarrow \omega
\end{aligned}
$$

De esta manera tenemos que la representación en serie de senos y cosenos de la función, ha quedado como una representación integral de exponenciales complejas
$$
f(t)=\frac{1}{2 \pi} \int_{-\infty}^{\infty}\left[\int_{-\infty}^{\infty} f(t) e^{-j \omega t} d t\right] e^{j \omega t} d \omega
$$

De donde se extrae el famoso valor de la transformada de Fourier:
$$
F(\omega)=\mathcal{F}[f(t)]=\int_{-\infty}^{\infty} f(t) e^{-j \omega t} d t
$$

Los resultados y las aplicaciones de esta transformada son múltiples. Uno de los resultados es que si tenemos, por ejemplo, la intensidad que recorre una malla de un circuito, esta se define en el dominio del tiempo (DT), mientras que la transformada de la intensidad se define en el dominio de la frecuencia (DF)
